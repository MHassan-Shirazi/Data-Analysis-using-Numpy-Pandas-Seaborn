```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Bank-Dataset.csv')
```

```python
df.drop(columns=['contact'], inplace = True)
df.drop(columns=['day'], inplace = True)
df.drop(columns=['duration'], inplace = True)
df.drop(columns=['campaign'], inplace = True)
df.drop(columns=['previous'], inplace = True)
df.drop(columns=['pdays'], inplace = True)
df.drop(columns=['poutcome'], inplace = True)
df.drop(columns=['default'], inplace = True)
```

```python
df['housing'] = df['housing'].astype(bool)
df['loan'] = df['loan'].astype(bool)
df['marital'] = df['marital'].astype(bool)
df['deposit'] = df['deposit'].astype(bool)
```

```python
sns.set_theme(style="whitegrid")
```

```python
# Cell 4 — Age Distribution
sns.histplot(df['age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Cell 5 — Age vs Deposit
sns.boxplot(x='deposit', y='age', data=df)
plt.title("Age vs Deposit")
plt.show()
```

```python
# Cell 6 — Balance Distribution (Log)
sns.histplot(np.log1p(df['balance']), kde=True)
plt.title("Balance Distribution (Log Scale)")
plt.show()
```

```python
# Cell 8 — Job vs Deposit
plt.figure(figsize=(12,6))
sns.countplot(x='job', hue='deposit', data=df)
plt.xticks(rotation=45)
plt.title("Job vs Deposit")
plt.show()
```

```python
# Cell 9 — Education vs Deposit
sns.countplot(x='education', hue='deposit', data=df)
plt.title("Education vs Deposit")
plt.show()
```

```python
# Cell 14 — Conversion Rate by Feature
def plot_conversion(feature):
    df.groupby(feature)['deposit'].mean().sort_values().plot(kind='bar')
    plt.title(f"Conversion Rate by {feature}")
    plt.ylabel("Deposit Rate")
    plt.show()

plot_conversion('job')
plot_conversion('education')
plot_conversion('marital')
plot_conversion('month')
plot_conversion('housing')
plot_conversion('loan')
```

```python
# Cell 16 — Scatter Plot (Age vs Balance)
sns.scatterplot(x='age', y='balance', hue='deposit', data=df)
plt.title("Age vs Balance vs Deposit")
plt.show()
```
