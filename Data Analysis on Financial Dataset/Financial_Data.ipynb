```python
df.drop(columns=['Symbol'], inplace = True)
```

```python
df.drop(columns=['Dividend Yield'], inplace = True)
df.drop(columns=['SEC Filings'], inplace = True)
```

```python
cols = [
    'Price','Price/Earnings','Earnings/Share',
    '52 Week Low','52 Week High',
    'Market Cap','EBITDA',
    'Price/Sales','Price/Book'
]

for col in cols:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(',', '', regex=False)              # remove commas
        .str.replace('$', '', regex=False)              # remove $
        .str.replace('B', 'e9', regex=False)            # 1.5B → 1.5e9
        .str.replace('M', 'e6', regex=False)            # 2.3M → 2.3e6
        .str.replace('K', 'e3', regex=False)            # 4.2K → 4.2e3
        .str.replace(r'[^0-9eE\.\-]', '', regex=True)   # remove garbage
        .replace('', np.nan)
    )

    df[col] = pd.to_numeric(df[col], errors='coerce')  # convert safely
```

Let's start by visualizing the distribution of the `Price` column using a histogram and Kernel Density Estimate (KDE) plot.

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Price'], kde=True, bins=30)
plt.title('Distribution of Stock Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

Next, let's visualize the distribution of companies across different `Sector` categories using a count plot.

```python
plt.figure(figsize=(12, 7))
sns.countplot(y='Sector', data=df, order=df['Sector'].value_counts().index, palette='viridis', hue='Sector', legend=False)
plt.title('Number of Companies per Sector')
plt.xlabel('Number of Companies')
plt.ylabel('Sector')
plt.grid(axis='x', alpha=0.75)
plt.show()
```

Let's also look at the distribution of 'Market Cap'.

```python
plt.figure(figsize=(12, 7))
sns.countplot(y='Sector', data=df, order=df['Sector'].value_counts().index, palette='viridis', hue='Sector', legend=False)
plt.title('Number of Companies per Sector')
plt.xlabel('Number of Companies')
plt.ylabel('Sector')
plt.grid(axis='x', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Market Cap'], kde=True, bins=30)
plt.title('Distribution of Market Capitalization')
plt.xlabel('Market Cap')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
